```python
def get_images(from_folder_list,buddha_image_list):
    
    cd = os.getcwd()
    
    for i in from_folder_list:
        
        if i.startswith('.'):
            continue
        else:
            for j in os.listdir(os.path.join(cd,i)):
                if j.startswith('.'):
                    continue
                else:
                    im = cv2.imread(os.path.join(cd,i,j))
                    print(j)
                    buddha_image_list.append(im)
         
    return buddha_image_list
```

```python
def resize_and_regroup(bil, to_folder):
    cd = os.getcwd()
    count = 0
    os.chdir(to_folder)
    for i in bil:
        im = cv2.resize(i, (256, 256))
        
        cv2.imwrite(str(count)+'.jpg', im)
   
        count = count + 1
    os.chdir(cd)
    return
        
        
```
